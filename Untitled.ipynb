```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

#sqlalchemy==1.4.1
#diffusers==0.12.1
#transformers
# librosa==0.8.1

from diffusers import PNDMScheduler
import os
os.chdir('/home/ray/default/')
         
from image_to_image import StableDiffusionImg2ImgPipeline
# from diffusers import StableDiffusionImg2ImgPipeline
import torch
from scipy.io.wavfile import read
from IPython.display import Audio
import seaborn as sns
import numpy as np
from PIL import Image

from uberduck_ml_dev.models.tacotron2 import DEFAULTS
from uberduck_ml_dev.models.common import MelSTFT

def spectrogram_from_image(
    image,
    power = 0.25,
    max_value = 10.,
) -> np.ndarray:

    if image.mode in ("P", "L"):
        image = image.convert("RGB")


    image = image.transpose(Image.Transpose.FLIP_TOP_BOTTOM)
    image = image.transpose(Image.FLIP_TOP_BOTTOM)
    data = np.array(image).transpose(2, 0, 1)
    data = data[0:1, :, :]
    data = data.astype(np.float32)
    data = 255 - data
    data = data / 255
    data = np.power(data, 1 / power)
    data = data * max_value

    return data


def image_from_spectrogram(spectrogram: np.ndarray, power: float = 0.25) -> Image.Image:
    """
    Compute a spectrogram image from a spectrogram magnitude array.
    This is the inverse of spectrogram_from_image, except for discretization error from
    quantizing to uint8.
    Args:
        spectrogram: (channels, frequency, time)
        power: A power curve to apply to the spectrogram to preserve contrast
    Returns:
        image: (frequency, time, channels)
    """
    # Rescale to 0-1
    max_value = np.max(spectrogram)
    data = spectrogram / max_value

    # Apply the power curve
    data = np.power(data, power)

    # Rescale to 0-255
    data = data * 255

    # Invert
    data = 255 - data

    # Convert to uint8
    data = data.astype(np.uint8)

    # Munge channels into a PIL image
    if data.shape[0] == 1:
        # TODO(hayk): Do we want to write single channel to disk instead?
        image = Image.fromarray(data[0], mode="L").convert("RGB")
    elif data.shape[0] == 2:
        data = np.array([np.zeros_like(data[0]), data[0], data[1]]).transpose(1, 2, 0)
        image = Image.fromarray(data, mode="RGB")
    else:
        raise NotImplementedError(f"Unsupported number of channels: {data.shape[0]}")

    # Flip Y
    image = image.transpose(Image.Transpose.FLIP_TOP_BOTTOM)

    return image



def image_from_spectrogram_sam(spectrogram: np.ndarray, power: float = 0.25) -> Image.Image:

    data = np.power(spectrogram, power)
    data = data * 255 / 50
    data = 255 - data
    data = np.asarray(data).astype(np.uint8)
    image = Image.fromarray(data[0], mode="L").convert("RGB")
    image = image.transpose(Image.Transpose.FLIP_TOP_BOTTOM)

    return image



scheduler = PNDMScheduler(
    beta_start=0.00085, beta_end=0.012, beta_schedule="scaled_linear"
)

stft = MelSTFT(
    filter_length=DEFAULTS.filter_length,
    hop_length=DEFAULTS.hop_length,
    win_length=DEFAULTS.win_length,
    n_mel_channels=DEFAULTS.n_mel_channels,
    sampling_rate=DEFAULTS.sampling_rate,
    mel_fmin=DEFAULTS.mel_fmin,
    mel_fmax=DEFAULTS.mel_fmax,
    padding=None,
)
```
